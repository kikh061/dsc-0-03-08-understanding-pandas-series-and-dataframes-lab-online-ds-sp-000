# Rename all the columns to lower case

```python
#Your code here
df = df.rename(str.lower, axis='columns')
df.head(6)
```

# Change the Index to be the Line Names

```python
#Your code here
df = df.set_index('linename')
df.head()
```

# Painstakingly change the index back

```python
# Your code here
df = df.reset_index()
df.head()
```

## Create another column 'Num_Lines' that is a count of how many lines pass through a station. Then sort your dataframe by this column in descending order.

```python
# Your code here
df['Num_Lines'] = df.linename.map(lambda x: len(x))
df.head()
```

## Write a function to clean a column name.

```python
def clean(col_name):
    cleaned = col_name.strip()
    cleaned = cleaned.title()
    return cleaned
```

```python
#This is a list comprehension. It applies your clean function to every item in the list.
#We then reassign that to df.columns
#You shouldn't have to change anything here.
#Your function above should work appropriately here.
df.columns = [clean(col) for col in df.columns] 
```

```python
#Checking the output, we can see the results.
df.columns
```

## Compare subway traffic by day of the week. Display this as a graph.

```python
#df.Date.dtype
#Your code here
#df.Date = pd.to_datetime(df.Date)
#print(df.Date.dtype)
#Preview updated dataframe
#df.head()
df['weekday'] = df['Date'].dt.weekday_name
#df.head()
import matplotlib.pyplot as plt

daily_riders =[]
daily_riders = df.groupby('weekday').size()
#print(daily_riders)
#plt.bar(daily_riders)
#plt.show()

#daily_riders.plot.bar()
grouped = df.groupby('weekday').sum()
grouped.plot(kind='barh')
grouped.head()
grouped = grouped.reset_index()
grouped.head()

#plt.bar(y, x, align='center')
#plt.ylabel('Number of Riders')
#plt.title('Number of Riders per Day')
 
#plt.show()
```

## Is there more subway traffic on a weekend or a weekday?    Be specific in comparing magnitudes.

```python
grouped['IsWeekend'] = grouped.weekday.map({'Monday':False,'Tuesday':False,'Wednesday':False,'Thursday':False,'Friday':False,'Saturday':True,'Sunday':True})
wkend = grouped.groupby('IsWeekend').mean()
wkend[['Entries', 'Exits']].plot(kind='barh')
```
